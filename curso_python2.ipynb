Calculando cuanto quita el soldado, monstruo y jefe final:
----------------------------------------------------------------

Asignamos cuanto quita el soldado:


```python
soldado = 1
```

Ahora un monstruo:

```python
monstruo = 3
```

Y por ultimo, el jefe final:

```python
jefeFinal = 5
```

Oh, y obviamente el jugador:

```python
jugador = 10
```

Bueno, pues actualmente el jugador tiene 10 vidas, y el soldado quita 1, el monstruo 3, y el jefe final 5, vamos a ver cuantas vidas nos quedan si nos atacan todos a la vez

```python
ataqueTotal = soldado + monstruo + jefeFinal
```

Si queremos sumar, restar etc... algo a una variable, en vez de hacer: variable = variable - variable 2 . Podemos poner el signo para el cálculo y continuamos con un igual, así: 

```python
jugador -= ataqueTotal
```

Y si no queremos hacer una variable para la suma total, le restamos la suma de todos los ataques en el mismo calculo:

```python
jugador -= soldado + monstruo + jefeFinal
```

Hey, no nos olvidemos que con tantas restas, no nos dará el valor original restado 1 vez, pongamos las vidas originales:

```python
jugador = 10
```

Y ahora si restamos solo una vez:

```python
jugador -= soldado + monstruo + jefeFinal
```
